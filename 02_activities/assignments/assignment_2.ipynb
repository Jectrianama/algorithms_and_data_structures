  # Question Two: Path to Leaves

  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  ### Example 1

  ![](./images/q1_ex1.png)

  Input: `root = [1, 2, 2, 3, 5, 6, 7]` *What traversal method is this?*

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
from typing import List
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        


def bt_path(root: TreeNode) -> List[List[int]]:
    result = []

    def dfs(node, path):
        if not node:
            return
        
        #Adding actual value to the path
        path.append(node.val)
        
        # if a leave
        if not node.left and not node.right:
            result.append(list(path))  
        
        # if not a leave, keep looking
        dfs(node.left, path)
        dfs(node.right, path)

        # go back
        path.pop()

    dfs(root, [])
    return result

# building tree example 1
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(2)
root.left.left = TreeNode(3)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

print(bt_path(root))

# building tree example 2
root = TreeNode(10)
root.left = TreeNode(9)
root.right = TreeNode(7)
root.left.left = TreeNode(8)

print(bt_path(root))
```

```python
'''
Given a list of numbers, the goal is to move all the zeros to the end of the list while keeping the order of the other numbers exactly the same.
This means you don’t sort the list or remove any numbers — you just shift the zeros to the back and keep everything else in its original order.
'''
```


-   Create 1 new example that demonstrates you understand the problem. Trace/walkthrough 1 example that your partner made and explain it.


```python
''' 
Foram's Example 1:
Input: nums = [7, 0, 0, 5, 0, 9]  
Goal: Move all the zeros to the end without changing the order of the other numbers

Steps:
1.	Start with the first number: 7 → keep it
2.	Next is 0, skip it for now
3.	Another 0, skip it too
4.	Then comes 5, place it after 7
5.	Next is 0, skip
6.	Last is 9, place it after 5

Now we have the non-zero numbers in order: [7, 5, 9]

After that, we add all the skipped zeros to the end: [0, 0, 0]

Final result: [7, 5, 9, 0, 0, 0]
'''
```


-   Copy the solution your partner wrote. 


```python
def move_zeros_to_end(nums: List[int]) -> List[int]:
    # Pointer to place the next non-zero element
    non_zero_index = 0

    # Move all non-zero elements to the front
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero_index], nums[i] = nums[i], nums[non_zero_index]
            non_zero_index += 1

    return nums
```


-   Explain why their solution works in your own words.


```python
''' 
This solution works because the for loop iterates through each number in the list. Every time it finds a non-zero, it swaps it with the number at the non_zero_index
(which starts at 0), making sure all non-zeros are pushed to the front in order.
''' 
```


-   Explain the problem’s time and space complexity in your own words.


```python
''' 
- The time complexity is O(n) since the algorithm iterates over the list once, with n being the total number of elements. 
- The space complexity is O(1) because the problem is solved directly within the original data, requiring no additional data structures or memory.
''' 
```
